```python
accesos_velocidad_provincia = pd.read_csv('accesos_velocidad_provincia.csv')
accesos_velocidad_provincia.info()
```

```python
accesos_velocidad_provincia.shape
```

```python
accesos_velocidad_provincia.head(10)
```

```python
accesos_velocidad_provincia['Provincia'] = accesos_velocidad_provincia['Provincia'].replace({'Capital Federal': 'Caba', 'Córdoba': 'Cordoba'})
```

```python
# Reemplazar los valores no finitos por ceros
accesos_velocidad_provincia['Año'] = accesos_velocidad_provincia['Año'].fillna(0)
# Convertir la columna 'Año' a enteros
accesos_velocidad_provincia['Año'] = accesos_velocidad_provincia['Año'].astype(int).astype(str)
```

```python
# Reemplazar los valores no finitos por ceros
accesos_velocidad_provincia['Trimestre'] = accesos_velocidad_provincia['Trimestre'].fillna(0)
# Convertir la columna 'Trimestre' a enteros
accesos_velocidad_provincia['Trimestre'] = accesos_velocidad_provincia['Trimestre'].astype(int).astype(str)
```

```python
accesos_velocidad_provincia.head(5)
```

```python
accesos_velocidad_provincia = accesos_velocidad_provincia.loc[(accesos_velocidad_provincia['Año'] != 0) & (accesos_velocidad_provincia['Trimestre'] != 0)]
```

```python
accesos_velocidad_provincia.head(5)
```

```python
accesos_velocidad_provincia = pd.read_excel('accesos_velocidad_provincia.xlsx')
```

```python
import numpy as np

# Definir los intervalos y etiquetas
bins = [0, 5, 20, 30, 50, 86, np.inf]
labels = ['muy lento', 'lento', 'normal', 'medio', 'rápido', 'muy rápido']

# Discretizar las columnas de velocidades de acceso
for col in accesos_velocidad_provincia.iloc[:, 4:]:
    accesos_velocidad_provincia[f'{col}_intervalos'] = pd.cut(accesos_velocidad_provincia[col], bins=bins, labels=labels)

# Mostrar los resultados
print(accesos_velocidad_provincia.head())
```

```python
accesos_velocidad_provincia.info()
```

```python

# seleccionar solo las columnas que contengan "Mbps" en el título
mbps_columns = [col for col in accesos_velocidad_provincia.columns if 'Mbps' in col]

# sumar los valores de cada columna
column_sums = accesos_velocidad_provincia[mbps_columns].sum()

# sumar los resultados de cada columna
total_sum = column_sums.sum()

# mostrar el resultado final
print(f"La suma total de las columnas es: {total_sum}")
```

## PENETRACION

```python
internet_penetracion=pd.read_csv ('datasets/penetracion/Internet_Penetracion.csv')
internet_penetracion.info()
```

```python
internet_penetracion['Provincia'] = internet_penetracion['Provincia'].replace({'Capital Federal': 'Caba', 'Córdoba': 'Cordoba'})
```

```python

# Reemplazamos las comas por puntos en la columna "Accesos por cada 100 hogares"
internet_penetracion['Accesos por cada 100 hogares'] = internet_penetracion['Accesos por cada 100 hogares'].str.replace(',', '.')

# Convertimos la columna "Accesos por cada 100 hogares" de tipo "object" a tipo "float"
internet_penetracion['Accesos por cada 100 hogares'] = internet_penetracion['Accesos por cada 100 hogares'].astype(float)
```

```python
internet_penetracion.head(10)
```

```python
internet_penetracion['fecha'] = pd.to_datetime(internet_penetracion['Año'].astype(str) + '-Q' + internet_penetracion['Trimestre'].astype(str))
```

```python
internet_penetracion.head(5)
```

```python
# Definimos una función lambda que reemplaza los puntos por comas en un valor
replace_dot_with_comma = lambda x: str(x).replace('.', ',')

# Aplicamos la función lambda a cada valor de la columna "Accesos por cada 100 hogares"
internet_penetracion['Accesos por cada 100 hogares'] = internet_penetracion['Accesos por cada 100 hogares'].apply(replace_dot_with_comma)
```

```python
print(internet_penetracion['Accesos por cada 100 hogares'].dtype)
```

```python
internet_penetracion.to_csv('internet_penetracion.csv', index=False)
```
